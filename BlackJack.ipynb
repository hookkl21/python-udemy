```python
#Game set up Global Variables
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card:
    
    def __init__(self,suit,rank):
        #assign global variables into attribute
        self.suit = suit
        self.rank = rank
        #since we are using dictionary to retrieve int we need to assign value as int
        #for later in the project to script the functions
        self.value = values[rank]
    
    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

```python
class Deck:
    
    def __init__(self):
        #create empty list to add elements into list as deck
        self.deck = []
        #iterate through the Card class attributes suit/rank
        for suit in suits:
            for rank in ranks:
                #adding each elements into the end of new empty list
                self.deck.append(Card(suit,rank))
    
    # This is for test to print one of the index of new deck list to see
    #if we created the deck correctly
    def __str__(self):
        return f"you have created a deck {self.deck}"
    
    #shuffling the deck using random.shuffle (imported random)
    def shuffle(self):
        random.shuffle(self.deck)
    
    #dealing one card at a time which means you need to pop/remove the end of the list's element
    def deal(self):
        return self.deck.pop()
```

```python
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of
    
    def add_card(self,card):
        #add one card to end of the list
        self.cards.append(card)
        #assign value defined by rank using dictionary (global variable: values)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1
    
    def adjust_for_ace(self):
        #if value over 21 AND has ace, ace can be transfer to 1 (11-10)
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:
    
    def __init__(self):
        # Simple default Value for player tip
        self.total = 100  
        self.bet = 0
        
    #if player/host wins, they gain the chip
    def win_bet(self):
        self.total += self.bet
    
    #if player/host lose, they lose the chip
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input("How much you wanna bet? "))
        except:
            print("Put an integer number!")
        else:
            if chips.bet > chips.total:
                print("Bet cannot exceed your total chips")
            else:
                break
```

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()   
```

```python
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while True:
        check = input("Would you like to hit or stay? Hit or stay").lower()
        
        if check.startswith("h"):
            hit(deck,hand)
        elif check.startswith("s"):
            print("Player stays, dealer's turn")
            playing = False
        else:
            print("Sorry, please try again.")
            continue
        break
```

```python
while True:
    # Print an opening statement
    print("Welcome to the BlackJack!")
    print("Your goal is to make cards closer to 21, if you are going over 21, you lose!")
    print("Dealer will hits until it reachs 17.")
    print("Ace can be 11 or 1 depends on your total number!")
    
    # Create & shuffle the deck, deal two cards to each player
    new_deck = Deck()
    new_deck.shuffle()
    
    player_hand = Hand()
    player_hand.add_card(new_deck.deal())
    player_hand.add_card(new_deck.deal())
    
    dealer_hand = Hand()
    dealer_hand.add_card(new_deck.deal())
    dealer_hand.add_card(new_deck.deal())
    
        
    # Set up the Player's chips
    player_chip = Chips()
    
    
    # Prompt the Player for their bet
    take_bet(player_chip)
    
    # Show cards (but keep one dealer card hidden)
    show_some(player_hand,dealer_hand)

    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(new_deck,player_hand)
        
        
        # Show cards (but keep one dealer card hidden)
        show_some(player_hand,dealer_hand)
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chip)
            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player_hand.value < 21:
        
        while dealer_hand.value < 17:
            hit(new_deck,dealer_hand)
        
        # Show all cards
        show_all(player_hand,dealer_hand)
        # Run different winning scenarios
        if  dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chip)
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chip)
        elif player_hand.value > dealer_hand.value:  
            player_wins(player_hand,dealer_hand,player_chip)
        else:
            push(player_hand,dealer_hand)
            
        
    
    # Inform Player of their chips total
    print("Player's total chips are: ", player_chip.total)
    
    # Ask to play again
    new_game = input("Would you like to play another hand? Enter 'y' or 'n' ")
    
    if new_game[0].lower()=='y':
        playing=True
        continue
    else:
        print("Thanks for playing!")
        break
```
